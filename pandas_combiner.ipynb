# Combine Data and Upload to MongoDB

## Import data job Github job postings API and Weather summary scrape

```python
#allow 2+ minutes to run

import pandas as pd

from jobs import cities_list, clean_df
from weather import weather_list
```

## Create dataframe from weather info

```python
weather_df = pd.DataFrame(weather_list)
weather_df
```

```python
clean_df.head()
```

## Merge job posting data with weather information

```python
new_df = clean_df.merge(weather_df,left_on="split_city",right_on="city")
new_df.head()
```

# Import Cost of Living info from scraping file

```python
from cost_of_living import COL_dict
```

```python
COL_df = pd.DataFrame(COL_dict)
COL_df.head()
```

## Merge Job+Weather data with Cost of Living data

```python
total_df = new_df.merge(COL_df,how="left",on="split_city")
total_df.head()
```

## Reduce to necessary columns

```python
final_df = total_df[["creation_date", "title", "company", "location", "description", "url", "weather", "COL Index", "Rent Index", "Groceries Index", "Restaurant Price Index", "Local Purchasing Power Index"]]
final_df.head()
```

## Convert df to dictionary list

```python
dict_for_mongo = final_df.to_dict("records")
dict_for_mongo
```

## Upload final data to MongoDB

```python
from pymongo import MongoClient
```

```python
# Create connection to MongoDB
client = MongoClient('localhost', 27017)
db = client['etl_test']

#JOBS
collection = db['final']
# Build a basic dictionary
d = dict_for_mongo
# Insert the dictionary into Mongo
for document in d:
    collection.insert(document)
```
